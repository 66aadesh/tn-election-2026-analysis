```python
import pandas as pd
from shapely import wkt

# Load your kaggle file
df = pd.read_csv(r"D:\TAMIL NADU PROJECT\RAW FILES\COORDINATES\assembly-constituency-map.csv")

# Convert geometry column to shapely objects
df["geometry"] = df["geometry"].apply(wkt.loads)

# Extract centroid lat and long
df["LAT"] = df["geometry"].apply(lambda x: x.centroid.y)
df["LONG"] = df["geometry"].apply(lambda x: x.centroid.x)

# Keep only what you need
output = df[["Constituency_No", "Constituency", "LAT", "LONG"]]

# Rename to match your format
output.columns = ["CONSTITUENCY_NO", "CONSTITUENCY", "LAT", "LONG"]

# Save
output.to_csv(r"D:\TAMIL NADU PROJECT\tn_centroids.csv", index=False)

print(f"Done. {len(output)} constituencies extracted.")
print(output.head())
```
