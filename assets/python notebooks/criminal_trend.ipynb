```python
import plotly.graph_objects as go

years = [2016, 2021, 2026]
all_criminal = [33, 60, 54]
serious_criminal = [18, 26, 24]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=years,
    y=all_criminal,
    mode="lines+markers+text",
    name="Declared Criminal Cases",
    line=dict(color="#378ADD", width=3),
    marker=dict(size=10),
    text=[f"{v}%" for v in all_criminal],
    textposition="top center",
    textfont=dict(size=13)
))

fig.add_trace(go.Scatter(
    x=years,
    y=serious_criminal,
    mode="lines+markers+text",
    name="Serious Criminal Cases",
    line=dict(color="#D85A30", width=3, dash="dash"),
    marker=dict(size=10),
    text=[f"{v}%" for v in serious_criminal],
    textposition="bottom center",
    textfont=dict(size=13)
))

fig.update_layout(
    title=dict(
        text="Winning Candidates with Declared Criminal Cases<br><sup>Source: ADR/myneta.info — self-declared ECI affidavit data</sup>",
        font=dict(size=16)
    ),
    xaxis=dict(
        tickvals=[2016, 2021, 2026],
        title="Election Year",
        tickfont=dict(size=13)
    ),
    yaxis=dict(
        title="% of Winning Candidates",
        range=[0, 75],
        ticksuffix="%",
        tickfont=dict(size=13)
    ),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.25,
        xanchor="center",
        x=0.5
    ),
    paper_bgcolor="white",
    plot_bgcolor="white",
    width=1000,
    height=550,
    font=dict(family="Arial")
)

fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=True, gridcolor="#F1EFE8")

import plotly.io as pio
pio.write_html(fig, r"D:\TAMIL NADU PROJECT\criminal_trend.html")
fig.show()
```
